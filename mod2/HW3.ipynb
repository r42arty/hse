```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


x = np.arange(0,100)
y = x*2
z = x**2
```

#### 1 Упражнение

```python
# Построение графика
plt.figure(figsize=(8, 6))
plt.plot(x, y)

# Настройки графика
plt.title("title")
plt.xlabel("x label")
plt.ylabel("y label")

# Убираем сетку
plt.grid(False)

# Добавляем внешнюю рамку
plt.gca().spines['top'].set_visible(True)
plt.gca().spines['right'].set_visible(True)
plt.gca().spines['bottom'].set_visible(True)
plt.gca().spines['left'].set_visible(True)

# Отображение
plt.show()
```

#### 2 Упражнение

```python
# Создание области с двумя графиками
fig, axs = plt.subplots(1, 2, figsize=(6, 4))

# Первый график: y = x * 2
axs[0].plot(x, y, 'b--', linewidth=3)  # Синий пунктир, толщина линии 3

# Второй график: z = x^2
axs[1].plot(x, z, 'r-', linewidth=3)  # Красная линия, толщина линии 3

# Настройки осей
for ax in axs:
    ax.grid(False)  # Без сетки
    ax.spines['top'].set_visible(True)
    ax.spines['right'].set_visible(True)

# Отображение
plt.tight_layout()
plt.show()
```

#### 3 Упражнение

```python
# Основной график
fig, ax = plt.subplots(figsize=(9, 6))
ax.plot(x, z, label="z = x^2")
ax.set_xlabel("X")
ax.set_ylabel("Z")

# Убираем сетку на основном графике
ax.grid(False)

# Вставка с увеличенной областью
inset_ax = fig.add_axes([0.3, 0.5, 0.3, 0.3])  # [x, y, width, height]
inset_ax.plot(x, y, label="y = x*2")
inset_ax.set_xlim(20, 22)  # Ограничиваем диапазон
inset_ax.set_ylim(30, 100)
inset_ax.set_title("zoom")
inset_ax.set_xlabel("X")
inset_ax.set_ylabel("Y")

# Убираем сетку на вставке
inset_ax.grid(False)

# Отображение
plt.show()
```
