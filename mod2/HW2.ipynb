11.Создайте лямбда-функцию (функцию), которая будет использовать строки с заглавной буквы. (хинт .capitalize()) и пррименить к колонкам Mjob и Fjob + записать в исходный df

```python
csv_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'
df = pd.read_csv(csv_url)
df
```

```python
df['Mjob'] = df['Mjob'].apply(lambda x: x.capitalize() if isinstance(x, str) else x)
df['Fjob'] = df['Fjob'].apply(lambda x: x.capitalize() if isinstance(x, str) else x)
```

12. Создайте функцию с именем majority, которая возвращает логическое значение в новый столбец с именем legal_drinker (считайте, что возраст majority превышает 17 лет).

```python
def majority(age):
    return age > 17

df['legal_drinker'] = df['age'].apply(majority)
df
```

13. Создайте функцию, которая проверяет тип данных. Если тип данныз int, то умножаем на 10, если нет , то передаем значение как есть.

```python
def mult_if_int(x):
    if isinstance(x, int):
        return x * 10
    else:
        return x
```

17. Соедините два df данных по строкам (использовать concat) , но предварительно надо создать df из данных ниже (data1, data2)
