### <center> Task №1 </center>

Создайте следующую функцию **`solver(n)`**, которая принимает на вход некоторое число **`n`** и внутри нее:

> **- a.** Принимайте строчки на вход от пользователя (консоли) до тех пор, пока не получите слово "Стоп". Создайте список из всех полученных слов через цикл.
>
> **- a*.** Сделайте предыдущий пункт через списочное выражение.
>
> **- b.** Объедините все строки в списке в одну длинную строчку, разделив изначальные строчки запятыми с пробелом (['a', 'b'] -> 'a, b').
>
> **- c.** Удалите в получившемся тексте все символы, кроме чисел, поменяв удаленные символы на пробелы (регулярное выражение).
>
> **- d.** Разделите получившийся текст из чисел по пробелам, сделав список. Сделайте из каждого элемента списка тип данных целое число ('12 27' -> ['12', '27'] -> [12, 27])
>
> **- e.** Попробуйте посчитать остаток от деления **`n`** на каждое число из списка по очереди. Если вдруг происходит деление на ноль, просто обработайте ошибку и переходите к следующему числу (try-except с ZeroDivisionError).
>
> **- f.** Получите 15 случайных целых чисел в диапазоне от 1 до 100 (numpy). Передайте по очереди каждое из них в функцию **`solver`**.  Ввод можете считать одинаковым для каждого эксперимента, сохранив при первом запуске в глобальную переменную (global), или получать каждый раз заново.
>
> **- g.** Замерьте время работы каждого запуска и всех запусков вместе.
>
> **- h.** Запишите получившиеся результаты эксперимента со временем (через f-строки) в файл 'Task_1.txt'.
>
>
> **Пример ввода:**
>
> Мне сегодня, 10 июня, 5 лет
> 
> 1 раз в 4 месяца можно загадать до 3 желаний
>
> Cтоп

```python
import random
import numpy as np
import re
import time

numbers_random = np.random.randint(1, 101, 15)

def solver(n):
    start_time = time.time()
    user_input = []
    
    # task_a
    print("Введите данные, или напишите Стоп для окончания ввода: ")
    while True:
        line_input = input()
        if line_input.strip().lower() == "стоп":
            break
        user_input.append(line_input)

    # task_a*
    user_input = [line_input for line_input in user_input]
    
    # task_b
    text_long = ", ".join(user_input)
    
    # task_c
    text_numbers = re.sub(r"[^0-9\s]", "", text_long)

    # task_d
    text_numbers_joined = list(map(int, text_numbers.split()))
    #print(text_numbers_joined)
    
    # task_e
    result = []
    for num in text_numbers_joined:
        try:
            result.append(n % 0)
        except ZeroDivisionError:
            result.append(None)
    
    # task_f
    for num_rand in numbers_random:
        solver(num_rand)
        
    # task_g
    total_time = time.time() - start_time

    # task_h
    with open("Task_1.txt", "a") as file_result:
        file_result.write(f"Результаты по времени {total_time}")
    
solver(10)
```

> **- d.** Рассчитайте медианы для столбцов возраста, выживаемости и стоимости билета. Создайте словарь, где ключ - название столбца, а значение - его медиана.

```python
import pandas as pd
import matplotlib.pyplot as plt

# task_a
df = pd.read_csv('https://raw.githubusercontent.com/rogovich/Data/master/data/titanic/train.csv')
display(df.tail())

# task_b

for pclass in [1]:
    class_ages = df[df["Pclass"] == pclass]["Age"].dropna()
    plt.hist(class_ages, bins = 20)
plt.title("1 класс")
plt.xlabel("Age")
plt.ylabel("Freq")
plt.show()

for pclass in [2]:
    class_ages = df[df["Pclass"] == pclass]["Age"].dropna()
    plt.hist(class_ages, bins = 20)
plt.title("2 класс")
plt.xlabel("Age")
plt.ylabel("Freq")
plt.show()

for pclass in [3]:
    class_ages = df[df["Pclass"] == pclass]["Age"].dropna()
    plt.hist(class_ages, bins = 20)
plt.title("3 класс")
plt.xlabel("Age")
plt.ylabel("Freq")
plt.show()

# task_c
survived_by_cabin = df.groupby("Cabin")["Survived"].mean().sort_values()
print(survived_by_cabin)

# task_d
median_dict = {
    "Age": df["Age"].median(),
    "Survived": df["Survived"].median(),
    "Fare": df["Fare"].median()
}
print(median_dict)
```
