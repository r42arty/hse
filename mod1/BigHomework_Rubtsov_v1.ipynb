````python
names_input = []    # Инициализируем пустой список для хранения ФИО

# user_names_input(names_input) для ввода пользователем списка ФИО
def user_names_input(names_in):
    # Вводим строки от пользователя, пока не встретим слово "Стоп/стоп/СТОП"
    while True:
        name = input("Введите Фамилию, Имя, Отчество или 'Стоп' для завершения: ")
        if name.lower() == "стоп":
            break
        names_in.append(name)
    return names_in

# format_name(full_name) для преобразования ФИО в формат "Фамилия И. О."
def format_name(full_name):
    parts = full_name.split()  # Разделяем строку на части (Фамилия Имя Отчество и прочие патронимы)
    last_name = parts[0]       # Фамилия остается полной
    initials = " ".join([f"{p[0]}." for p in parts[1:]])  # Инициалы имени, отчества и прочих патронимов
    return f"{last_name} {initials}"

# task1() основная функция по заданию 1
def task1():
    user_names_input(names_input) # Запускаем ввод ФИО

    # Используем lambda-функцию, которая в каждой строке все слова сделает с большой буквы
    names_formatted = list(map(lambda x: x.title(), names_input))

    # Применяем форматирование к каждому ФИО
    result = [format_name(name) for name in names_formatted]
    
    # Печатаем результат, соединяя Фамилию И.О. через запятую
    print(", ".join(result))


task1()
````

````python
import random   # Для создания массива из рандомных чисел

# timer(arg_func) для подсчета времени и количества вызовов
def timer(arg_func):
    import time
    call_time = 0   # Переменная для подсчета времени вызовов
    call_count = {} # Словарь для подсчета количества вызовов

# wrapper(*args, **kwargs) для подсчета количества вызовов
    def wrapper(*args, **kwargs):
        nonlocal call_time, call_count
        start = time.time()                 # Время начала
        result = arg_func(*args, **kwargs)  # Запуск функции 
        time_taken = time.time() - start    # Прошедшее время: время сейчас - время старта
        call_time += time_taken
        
        func_name = arg_func.__name__       # Получаем имя функции
        if func_name in call_count:         # Увеличиваем количество запусков для конкретной функции
            call_count[func_name] += 1
        else:
            call_count[func_name] = 1

        print(f'Taken time: {round(time_taken * 1000, 5)} ms, Total time: {round(call_time * 1000, 5)} ms, Call count: {call_count}')
        return result

    return wrapper

@timer  # Используем декоратор timer

# square_sum(left, right) возвращает сумму квадратов в диапазоне от left до right
def square_sum(left, right):
    return sum(i ** 2 for i in range(left, right + 1))

@timer # Используем декоратор timer

# take_greatest_even(n, k) сортирует случайный массив целых чисел размера n
# по убыванию и выводит первые k четных чисел
def take_greatest_even(n, k):
    # Генерация случайного массива из n целых чисел от 1 до 100
    random_array = random.sample(range(1, 101), n)
    
    # Отбор только четных чисел и сортировка их по убыванию
    evens = sorted([num for num in random_array if num % 2 == 0], reverse=True)
    return evens[:k]    # Возврат первых k четных чисел

# task2() основная функция по заданию 2
def task2():
    # Получаем данные от пользователя для функции square_sum (ввод в одной строке)
    left, right = map(int, input("Введите через пробел левую и правую границы диапазона: ").split())
    square_sum_result = square_sum(left, right)
    print(f"Сумма квадратов чисел в диапазоне от {left} до {right}: {square_sum_result}")

    # Получаем данные от пользователя для функции take_greatest_even (ввод в одной строке)
    n, k = map(int, input("Введите через пробел размер массива и количество четных чисел для вывода: ").split())
    greatest_even_result = take_greatest_even(n, k)
    print(f"Первые {k} четных чисел в отсортированном массиве размерностью {n}: {greatest_even_result}")


task2()
````

### <center> Task №4 </center>

> **Intro**: Представьте, что вы получили зашифрованное сообщение от друга, которое нужно расшифровать. 
>
> **- a.** Каждый символ в исходном сообщении заменяется на символ, который находится на **`n`** позиций вперед по алфавиту (например, если `n=3`, 'A' заменяется на 'D', 'B' заменяется на 'E', 'C' заменяется на 'F' и т.д.). **`n`** передается пользователем в начале программы. Алфавит цикличен, т.е. после 'Z' идет 'A'.
>
> **- b.** Создайте словарь, где ключ - буква в алфавите, а значение: его зашифрованная версия. Напишите функцию **`encode(text, dict)`**, которая возвращает зашифрованный текст. Используйте методы строк `ord()`, `chr()`, `find()`, `upper()`, `lower()`, `replace()` для реализации шифрования. Помните, что текст может быть в любом регистре. Старайтесь сделать эффективную реализацию.
>
> **- c.** Напишите функцию **`decode(текст, словарь)`**, которая принимает строку зашифрованного текста и возвращает исходный текст. Используйте те же методы строк, что и в функции **`encode()`**, для обратного преобразования. Не перепутайте ключи и значения - теперь вам предстоит создать обратный словарь или работать особенным способом с изначальным. 
>
> **- d.** Примите на вход строчку и команду. Если `команда == 'decode'`, то расшифруйте сообщение, иначе если `команда == 'encode'`, то зашифруйте. В противном случае выведите пользователю, что он все перепутал
>
> **Пример ввода:**
>
> 3
> 
> QEB NRFZH YOLTK CLU GRJMP LSBO QEB IXWV ALD, decode
>
> **Пример вывода:** 
>
> THE QUICK BROWN FOX JUMPS OVER THE LAZY DOG

````python
# Функция для создания словаря шифрования
def create_cipher_dict(n):
    alphabet_upper = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    alphabet_lower = "abcdefghijklmnopqrstuvwxyz"
    cipher_dict = {}

    # Создаем словарь для шифрования
    for i in range(len(alphabet_upper)):
        cipher_dict[alphabet_upper[i]] = alphabet_upper[(i - n) % 26]
        cipher_dict[alphabet_lower[i]] = alphabet_lower[(i - n) % 26]
    
    return cipher_dict

# encode(text, cipher_dict) для шифрования текста
def encode(text, cipher_dict):
    encrypted_text = ''
    for char in text:
        if char.isalpha():  # Шифруем только буквы
            encrypted_text += cipher_dict[char]
        else:
            encrypted_text += char  # Пробелы и знаки препинания остаются
    return encrypted_text

# decode(text, cipher_dict) для дешифрования текста
def decode(text, cipher_dict):
    reversed_dict = {v: k for k, v in cipher_dict.items()}  # Создаем обратный словарь
    decrypted_text = ''
    for char in text:
        if char.isalpha():  # Дешифруем только буквы
            decrypted_text += reversed_dict[char]
        else:
            decrypted_text += char  # Пробелы и знаки препинания остаются
    return decrypted_text

# task4() основная функция по заданию 4
def task4():
    try:
        n = int(input("Введите сдвиг n: "))  # Получаем сдвиг
        cipher_dict = create_cipher_dict(n)  # Создаем словарь шифрования

        # Получаем сообщение и команду от пользователя
        message = input("Введите сообщение: ")
        command = input("Введите команду (encode/decode): ").strip().lower()

        if command == 'encode':
            result = encode(message, cipher_dict)
            print(f"Зашифрованное сообщение: {result}")
        elif command == 'decode':
            result = decode(message, cipher_dict)
            print(f"Расшифрованное сообщение: {result}")
        else:
            print("Команда введена неверно. Используйте 'encode' или 'decode'.")
    except ValueError:
        print("Ошибка! Сдвиг должен быть числом.")


task4()
````

> **Intro**: Магический квадрат - это квадратная матрица, где сумма чисел в каждой строке, каждом столбце и каждой диагонали одинакова. 
>
> **- a.** Напишите функцию **`generate_matrix(n)`**, которая генерирует квадратную матрицу размером n x n, заполненную случайными числами от 1 до n^2. Например, для **`n=3`** функция должна сгенерировать матрицу размером 3x3 со случайными числами от 1 до 9. **`n`** вы получите, как всегда, от пользователя.
>
> **- b.** Напишите функцию **`is_magic_square(matrix)`**, которая принимает матрицу и проверяет, является ли она магическим квадратом. Функция должна вернуть `True`, если матрица - магический квадрат, и `False` в противном случае.
>
> **- c.** Используйте цикл while для генерации матриц размером nxn до тех пор, пока не будет найден магический квадрат. Выведите найденный магический квадрат. 
>
> **Пример ввода:**
>
> 3

````python
import random

# generate_matrix(n) для генерации квадратной матрицы размером n x n, заполненной случайными числами от 1 до n^2
def generate_matrix(n):
    numbers = random.sample(range(1, n**2 + 1), n**2)           # Генерируем случайные числа от 1 до n^2 без повторений
    matrix = [numbers[i * n:(i + 1) * n] for i in range(n)]     # Преобразуем список чисел в матрицу n x n
    return matrix

# is_magic_square(matrix) для проверки, является ли матрица магическим квадратом
# Магический квадрат — это матрица, где суммы чисел в каждой строке, столбце и на обеих диагоналях равны
def is_magic_square(matrix):
    size = len(matrix)  # Определяем размер матрицы (n)
    magic_sum = sum(matrix[0])  # Считаем сумму первой строки, с которой будем сравнивать остальные суммы

    # Проверяем, что суммы всех строк равны magic_sum
    for row in matrix:
        if sum(row) != magic_sum:
            return False

    # Проверяем, что суммы всех столбцов равны magic_sum
    for col in range(size):
        if sum(matrix[row][col] for row in range(size)) != magic_sum:
            return False

    # Проверяем, что сумма на главной диагонали равна magic_sum
    if sum(matrix[i][i] for i in range(size)) != magic_sum:
        return False

    # Проверяем, что сумма на побочной диагонали равна magic_sum
    if sum(matrix[i][size - i - 1] for i in range(size)) != magic_sum:
        return False

    return True # Если все проверки прошли — это магический квадрат


# task5() основная функция по заданию 5
def task5():
    # ввод размера матрицы nxn
    n = int(input("Введите размерность квадратной матрицы n (при n > 3 алгоритм будет отрабатывать долго): "))

    found_magic_square = False          # Флаг -- найден ли магический квадрат
    attempts = 0                        # Счетчик попыток
    while not found_magic_square:
        matrix = generate_matrix(n)     # Генерируем матрицу
        if is_magic_square(matrix):     # Проверяем, является ли она магическим квадратом
            found_magic_square = True
            print(f"Найден магический квадрат (попытка {attempts+1}):")
            for row in matrix:
                print(row)
        attempts += 1


task5()
````

Напишите функцию **`find_artefact(items)`**, которая использует `filter()` для отбора из списка `items` только те, что содержат слово "камень". Напишите функцию **`recover(people, artefacts)`**, которая принимает словарь подданных и найденные артефакты. Функция должна использовать `map()` и `lambda`-функцию, чтобы вернуть список восстановленных имен подданных (если найден "Камень Желаний" среди артефактов) или оставить имена животных (если артефакт не найден).

````python
from numpy.random import choice # Импортируем только функцию choice из numpy

# find_artefact(items) для фильтрации предметов, которые содержат слово "камень"
def find_artefact(items):
    return list(filter(lambda item: "камень" in item.lower(), items))

# check_artefact(items) для проверки, является ли предмет "Камень Желаний"
def check_artefact(items):
    return list(map(lambda item: item.lower() == "камень желаний", items))

# recover(people, artefacts) для восстановления подданных, если найден "Камень Желаний"
def recover(people, artefacts):
    found = any(artefacts)  # Проверяем, найден ли "Камень Желаний"
    
    if found:
        print("Камень Желаний найден! Подданные восстанавливаются в человеческий облик")
    else:
        print("Камень Желаний не найден! Подданные остаются животными")
    
    # Если найден, возвращаем имена, иначе — животных
    return list(map(lambda person: person.split(", ")[0] if found else person.split(", ")[1], people))


# task7() основная функция по заданию 7
def task7():
    # Получаем список предметов в лесу
    items = []
    while True:
        item = input("Введите название предмета, найденного в лесу (или 'Стоп' для завершения): ")
        if item.lower() == "стоп":
            break
        items.append(item)

    # Выбираем 5 случайных предметов из списка введенных пользователем (с повторами)
    random_items = choice(items, 5, replace=True)  # Выбор 5 случайных предметов с возможностью повторения
    print(f"Случайные предметы, найденные в лесу: {list(random_items)}")
    
    # Фильтруем предметы, чтобы найти те, что содержат "камень"
    filtered_items = find_artefact(random_items)
    print(f"Артефакты, содержащие 'камень': {filtered_items}")

    # Проверяем, есть ли среди найденных артефактов "Камень Желаний"
    artefact_check = check_artefact(filtered_items)
    print(f"Проверка артефактов на 'Камень Желаний': {artefact_check}")

    # Получаем список подданных, превратившихся в животных
    people = []
    while True:
        person = input("Введите <имя, животное> в которое превратился подданный (или 'Стоп' для завершения): ")
        if person.lower() == "стоп":
            break
        people.append(person)

    # Восстанавливаем подданных, если среди случайных предметов найден "Камень Желаний"
    restored_people = recover(people, artefact_check)
    print(f"Итоговые подданные: {restored_people}")


task7()
````
